- 1: Plot the function

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x, y: 5 * x**2 - 6 * x * y + 5 * y**2 + 4 * x + 4 * y

x = np.linspace(-6, 8, 200)
y = np.linspace(-6, 8, 200)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.plot_surface(X, Y, Z)
min = np.array([-1, -1])
ax.scatter(min[0], min[1], 0, c="r", s=100, marker="*")

f(min[0], min[1])
```

- 2: Compute analitically the minimum of the function

Given the function 
$$f(\mathbf{x}) = 5x^2 - 6xy + 5y^2 + 4x + 4y$$
I can find the minimum by computing the gradient and setting it to zero:
$$\nabla f(\mathbf{x}) = \begin{bmatrix} 10x - 6y + 4 \\ -6x + 10y + 4 \end{bmatrix} = \mathbf{0}$$
and obtain the solution
$$\mathbf{x_{min}} = \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -1 \\ -1 \end{bmatrix}$$
and the minimum value of the function
$$f(\mathbf{x_{min}}) = -4$$

- 6: Implement the gradient descent method.

```python
tol = 1.0e-3
n_iter = 200
xk = np.array([-5, 7])

gradientx = lambda x, y: 10 * x - 6 * y + 4
gradienty = lambda x, y: -6 * x + 10 * y + 4

learning_rate = 0.125
convergence_path = list()
convergence_path.append(xk.copy())
old_xk = xk.copy()
for i in range(n_iter):
    xk = xk - learning_rate * np.array(
        [gradientx(xk[0], xk[1]), gradienty(xk[0], xk[1])]
    )
    if np.abs(f(xk[0], xk[1]) - f(old_xk[0], old_xk[1])) < tol:
        break
    old_xk = xk.copy()
    convergence_path.append(xk.copy())

print("Last point: ", xk)
print("Number of iterations: ", i)

plt.contourf(X, Y, Z, 10)
plt.plot(
    [x[0] for x in convergence_path], [x[1] for x in convergence_path], "o-", c="r"
)
plt.xlim([-6, 8])
plt.ylim([-6, 8])
```

As we can see with learning rate set as the maximum possible value, the algorithm diverges.
